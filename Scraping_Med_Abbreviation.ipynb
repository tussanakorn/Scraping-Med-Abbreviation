```
import requests
import bs4
import pandas as pd
```

```
response = requests.get('https://www.medicinenet.com/common_medical_abbreviations_and_terms/article.htm')
```

```
print(response)
```

```
print(response.content)
```

```
print(type(response.content))
```

```
html_page = bs4.BeautifulSoup(response.content,'lxml')
```

```
print(html_page)
```

```
html_page.select('li strong') 
```

```
med_all = html_page.select('div.wrapper')
```

```
med_all
```

```
med_name_list = []
for i in range(4,len(med_all)):
  selects = med_all[i].select('li')
  for select in selects:
    select1 = select.text
    med_name_list.append(select1)
```

```
med_name_list
```

```
len(med_name_list)
```

```
#med_desc_list = []
for i in range(4,len(med_all)):
  selects = med_all[i].find_all('li')
  for select in selects:
  
    #select2 = select['href']
    print(select)
    
    # med_list_abb = select.text
    # medlist_all.append(med_list_abb)
```

```
med_link_list = []
for i in range(4,len(med_all)):
  selects = med_all[i].find_all('li')
  for select in selects:
    href = [i['href'] for i in select.find_all('a', href=True)]
    med_link_list.append(href)
```

```
med_link_list
```

```
len(med_link_list)
```

```
med_names = []
for name in med_name_list:
  name_abb = name.split(':')
  med_names.append(name_abb)
```

```
med_names
```

```
df = pd.DataFrame(med_names)
```

```
df.to_csv('Med_Abbrev.csv',index=False)
```
